```python
# Q4
class TreeNode:
    def __init__(self, initKey, initParent = None, initLeftChild = None, initRightChild = None):
        self.key = initKey
        self.parent = initParent
        self.leftChild = initLeftChild
        self.rightChild = initRightChild
        self.nw = 0


class BinarySearchTree:
    def __init__(self, initRootNode = None):
        self.root = initRootNode
        self.treesize = 0
        if initRootNode != None:
            self.treesize += 1

    # IMPLEMENT HERE
    # return the node with key value = k, otherwise return None
    def search(self, k, currNode):
        if currNode == None:
            return None
        if k < currNode.key:
            return self.search(k, currNode.leftChild)
        elif k > currNode.key:
            return self.search(k, currNode.rightChild)
        else:
            return currNode

    # IMPLEMENT HERE
    # create a new node with key value = k, and insert it into the tree, and return the newly created node
    # Note: nw of all affected nodes needs to be updated here
    def insert(self, k, currNode):
        if currNode == None:
            return
        if k < currNode.key:
            if currNode.leftChild == None:
                n = TreeNode(k)
                currNode.leftChild = n
                n.parent = currNode
                n.nw = 1
                itr = currNode
                while itr != None:
                    itr.nw += 1
                    itr = itr.parent
                return n
            else:
                self.insert(k, currNode.leftChild)
        elif k > currNode.key:
            if currNode.rightChild == None:
                n = TreeNode(k)
                currNode.rightChild = n
                n.parent = currNode
                n.nw = 1
                itr = currNode
                while itr != None:
                    itr.nw += 1
                    itr = itr.parent
                return n
            else:
                self.insert(k, currNode.rightChild)
        else:
            return currNode

    # IMPLEMENT HERE
    # delete the node with key value = k; do nothing if not found
    # Note: nw of all affected nodes needs to be updated here
    def delete(self, k):
        n = self.search(k, self.root)       
        if n == None:
            return
        if n.leftChild == None and n.rightChild == None:
            if n.parent != None:
                if n.parent.leftChild == n:
                    n.parent.leftChild = None
                else:
                    n.parent.rightChild = None
            else:
                self.root = None
        elif n.leftChild and n.rightChild:
            tmp = n.rightChild
            while tmp.leftChild != None:
                tmp = tmp.leftChild
            if tmp.parent.leftChild == tmp:
                tmp.parent.leftChild = tmp.rightChild
            else:
                tmp.parent.rightChild = tmp.rightChild
            if tmp.rightChild:
                tmp.rightChild.parent = tmp.parent
            tmp.parent = n.parent
            tmp.leftChild = n.leftChild
            if n.rightChild != tmp:
                tmp.rightChild = n.rightChild
            if n.leftChild:
                tmp.leftChild.parent = tmp
            if n.rightChild:
                tmp.rightChild.parent = tmp
            if n.parent:
                if n == n.parent.leftChild:
                    n.parent.leftChild = tmp
                else:
                    n.parent.rightChild = tmp
            else:
                self.root = tmp
        else:
            if n.leftChild != None:
                n.leftChild.parent = n.parent
                if n.parent:
                    if n.parent.leftChild == n:
                        n.parent.leftChild = n.leftChild
                    else:
                        n.parent.rightChild = n.leftChild
                else:
                    self.root = n.leftChild
            else:
                n.rightChild.parent = n.parent
                if n.parent:
                    if n.parent.leftChild == n:
                        n.parent.leftChild = n.rightChild
                    else:
                        n.parent.rightChild = n.rightChild
                else:
                    self.root = n.rightChild
        del n

    # IMPLEMENT HERE
    # search range [minimum, maximum] both inclusive; assume minimum <= maximum already guaranteed
    # return the list of key values in the range
    def searchByRange(self, minimum, maximum, currNode):
        if currNode == None:
            return []
        if currNode.key < minimum:
            return self.searchByRange(minimum, maximum, currNode.rightChild)
        elif currNode.key >= minimum and currNode.key <= maximum:
            return self.searchByRange(minimum, maximum, currNode.leftChild) + [currNode.key] + self.searchByRange(minimum, maximum, currNode.rightChild)
        else:
            return self.searchByRange(minimum, maximum, currNode.leftChild)
        
    # IMPLEMENT HERE
    # search and return the i-th smallest node, return None if i is out of range
    def searchith(self, i, currNode):
        if currNode == None:
            return None
        nw = 1
        if currNode.leftChild != None:
            nw = currNode.leftChild.nw + 1
        if i == nw:
            return currNode
        elif i < nw:
            return self.searchith(i, currNode.leftChild)
        else:
            return self.searchith(i-nw, currNode.rightChild)

    # print the tree layer by layer, in increasing order of depth       
    def printTree(self):
        x = [self.root]              
        print("--------- Tree begins here ----------")
        NonEmptyLevel = True
        while NonEmptyLevel == True:
            y = []
            NonEmptyLevel = False
            for i in range(len(x)):
                if x[i] != None:
                    print(x[i].key, end=' ')
                    NonEmptyLevel = True
                    y += [x[i].leftChild, x[i].rightChild]
                else:
                    print(' ', end=' ')
            x = y
            print()
        print("--------- Tree ends here -------------")

    
    def inorder(self, currNode):
        if currNode.leftChild:
            self.inorder(currNode.leftChild)
        print(currNode.key, currNode.nw)
        if currNode.rightChild:
            self.inorder(currNode.rightChild)
            
    def preorder(self, currNode):
        print(currNode.key, end='|')
        if currNode.leftChild:
            self.preorder(currNode.leftChild)
        if currNode.rightChild:
            self.preorder(currNode.rightChild)
            
    def preorder_next(self, key):
        # Return subsequent node of specified node in
        # preorder traversal of a proper binary tree
        # In a proper binary tree, if node is internal then we always return the left child
        currNode = self.search(key, currNode = self.root)
        if currNode.leftChild:
            return (currNode.leftChild).key
        # Otherwise the node is external, so we consider whether 
        # it is a left child or right child of its parent node
        if (currNode.parent).leftChild == currNode:
            # If node is a left child of its parent, then we return its sibling
            return ((currNode.parent).rightChild).key
        elif (currNode.parent).rightChild == currNode:
            # If node is a right child of its parent, then we traverse the parent's 
            # ancestors until one has a right child we have not encountered
            node = currNode.parent
            while node.parent:
                temp = node
                node = node.parent
                if node.rightChild == temp:
                    continue
                else:
                    return (node.rightChild).key
            else:
                # We have reached the root so return None as we reached end of traversal
                return None
                
    def inorder_next(self, key):
        # Return subsequent node of specified node in inorder 
        # traveral of a proper binary tree
        currNode = self.search(key, currNode = self.root)
        # If node is internal, then we recursively traverse its right subtree
        # until we find the left-most external node
        if currNode.leftChild:
            # Start traversing from right subtree
            currNode = currNode.rightChild
            while currNode.leftChild != None:
                currNode = currNode.leftChild
            else:
                return currNode.key
        # If node is external, we consider whether it is a left child or right child
        # If node is a left child of some parent node, we return the parent node
        if (currNode.parent).leftChild == currNode:
            return (currNode.parent).key
        elif (currNode.parent).rightChild == currNode:
            # If node is a right child of some parent node, then we traverse the ancestors
            # until one has a larger key value than node
            while currNode.parent:
                if currNode.key > (currNode.parent).key:
                    currNode = currNode.parent
                else:
                    return (currNode.parent).key
            else:
                # We have reached the root so return None as we reached end of traversal
                return None
                
    def postorder_next(self, key):
        # Return subsequent node of specified node in postorder 
        # traversal of proper binary tree
        currNode = self.search(key, currNode = self.root)
        # If node is root, then we return None since it is the last node to be visited
        if currNode == self.root:
            return None
        # Else we consider if node is left child or right child
        if (currNode.parent).leftChild == currNode:
            # If it is a left child, the next node is the left-most external node of its sibling
            node = (currNode.parent).rightChild    # Sibling node
            while node.leftChild:
                node = node.leftChild
            else:
                return node.key
        elif (currNode.parent).rightChild == currNode:
            # If it is a right child, the next node is its parent
            return (currNode.parent).key
        
# Complexity analysis for preorder_next, postorder_next and inorder_next:
# Worst case occurs when we are traversing from an external node of maximal depth
# back to the root and then to another external node of maximal depth.
# This takes in total O(h) + O(h) = O(h + h) = O(2h) time.
# So worst case time complexity would be O(h) time.
```

```python
# Q5

def computeDepth(v,current_depth): 
    # INPUT: a node v from a proper binary tree T 
    # OUTPUT: prints the depth of v and of all the nodes in his subtree
    print("node" + "is at depth %d" % current_depth) 
    if v.left != None: 
        computeDepth(v.left ,current_depth +1) 
    if v.right != None: 
        computeDepth(v.right ,current_depth +1)
```

```python
# Q6

def Euler_tour(v, T):
    # INPUT: a node v that is root from a tree T
    # OUTPUT: an Euler tour traversal of T
    
    # Print root node once
    print(v.key) 
    
    while True:
        while v.left != None:
            v = v.left
            print(v.key)
            
        # Once we reach external node, we need to print twice more
        print(v.key)
        print(v.key)
        
        # Once we reach an external right node, we need to traverse
        # backwards up to the root
        while (v.parent).right == v:
            v = v.parent
            print(v.key)
            
        # Go back to parent node and visit the right subtree
        v = v.parent
        print(v)
        v = v.right
        print(v.key)
        
```

```python
# Q7

class Stack:
    
    def __init__(self):
        self.array = []
        
    def push(self, k):
        self.array.append(k)
        
    def pop(self):
        return self.array.pop()
    
    def isEmpty(self):
        return len(self.array) == 0

def iter_inorder(v, T):
    # INPUT: a node v, root of a proper binary tree T 
    # OUTPUT: prints the nodes of T in a inorder traversal
    
    s = Stack()
    
    while True:
        
        # We keep going left, stack keeps track of nodes traversed
        while v.left:
            s.push(v)
            v = v.left
        
        # We reached an external node 
        print(v.key)
        
        # Exit condition 
        if s.isEmpty():
            return 
        
        # We kept ancestors of v in the stack
        v = s.pop()
        
        # Visit parent of v
        print(v.key)
        
        # Start traversing from the right subtree
        v = v.right
```
